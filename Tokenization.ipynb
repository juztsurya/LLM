```python
text = "Hello world! How are you?"
tokens = text.split()
print(tokens)  # ['Hello', 'world!', 'How', 'are', 'you?']
```

```python
text = "Hello"
tokens = list(text)
print(tokens)  # ['H', 'e', 'l', 'l', 'o']
```

```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
tokens = tokenizer.tokenize("Unhappiness")
print(tokens)  # ['un', '##happi', '##ness']
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True) # Added to download the missing resource
stop_words = set(stopwords.words('english'))
text = "This is a sample text with stop words"
tokens = [w for w in word_tokenize(text) if w.lower() not in stop_words]
print(tokens)  # ['sample', 'text', 'stop', 'words']
```

```python
import nltk
from nltk.tokenize import sent_tokenize
nltk.download('punkt', quiet=True)
text = "Hello world. This is a test."
tokens = sent_tokenize(text)
print(tokens)  # ['Hello world.', 'This is a test.']
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt', quiet=True)
text = "Hello, world!"
tokens = word_tokenize(text)
print(tokens)  # ['Hello', ',', 'world', '!']
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("Hello world!")
tokens = [token.text for token in doc]
print(tokens)  # ['Hello', 'world', '!']
```
